## 8.1 Iris
This dataset can be accessed through the sklearn package (https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html). It includes three iris species with 50 samples each as well as some properties about each flower. One flower species is linearly separable from the other two, but the other two are not linearly separable from each other. The columns in this dataset are: Id, SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm, Species.

```python
from sklearn.datasets import load_iris
```

```python
# load data
iris = load_iris()
iris = pd.concat([pd.DataFrame(iris['data'], columns = ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]),
                  pd.DataFrame(iris['target'], columns = ['Species'])], axis = 1)
print("This dataframe has {} rows and {} columns.".format(iris.shape[0], iris.shape[1]))
print(iris.head())
```

```python
# distribution of target variable
ax = sns.countplot(x="Species", data=iris).set_title('Distribution of Target Variable')
```

```python
# pairwise correlation between features
ax = sns.heatmap(iris[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]].corr(), vmin = -1, vmax=1, annot=True).set_title('Heatmap')
```

```python
# relationship of features and target variable
ax = sns.pairplot(iris, hue = "Species")
```

```python
# distribution of features by label
sns.catplot(x="Species", y="SepalLengthCm", kind="box", data=iris)
plt.title('Boxplot for SepalLength by Label')
plt.show()
```

## 8.3 Boston
This dataset can be accessed through the sklearn package (https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_boston.html). It includes data on housing prices in Boston.

```python
from sklearn.datasets import load_boston
```

```python
# load data
boston = load_boston()
boston = pd.concat((pd.DataFrame(boston["data"], columns = boston["feature_names"]), pd.DataFrame(boston["target"], columns=["target"])), axis = 1)
print("This dataframe has {} rows and {} columns.".format(boston.shape[0], boston.shape[1]))
print(boston.head())
```

```python
# distribution of target variable
ax =sns.distplot(boston["target"])
plt.title("Distribution of Target")
plt.show()
```

```python
# outlier analysis
plt.boxplot(boston["target"])
plt.title("Boxplot for Target")
plt.show()
```

```python
# pairwise correlation between features
fig, ax = plt.subplots()
fig.set_size_inches(15,8)
ax = sns.heatmap(boston.iloc[:,:-1].corr(), vmin = -1, vmax=1, annot=True).set_title('Heatmap')
```

## 8.5 Movements
This dataset includes daily stock movements on 60 S&P500 companies from 2010-01-04 until 2013-10-29.

```python
# load data
movements = pd.read_csv("datasets/Movements.csv")

# take a look at dataframe
print("This dataframe has {} rows and {} columns.".format(movements.shape[0], movements.shape[1]))
print(movements.head(1))
```

```python
# dataframe manipulation
movements_transp = movements.T.rename(columns = dict(movements.T.iloc[0,:])).iloc[1:,:]

# plot return
plt.rcParams["figure.figsize"] = (15,8)
movements_transp[["Amazon", "Bank of America", "Exxon", "Wal-Mart"]].plot()
plt.xlabel('Date')
plt.ylabel('Return %')
plt.title('Daily Returns of Selected Stocks')
plt.show()
```
